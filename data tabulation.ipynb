```python
import numpy as np
import pandas as pd
import os
import shutil
```

```python
years = ['2017','2019','2020']
common = 'NIFTY50_'
months = ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']
location = 'C:/Users/ajith/research/data/while_year_data/'
companies = ['HDFCBANK',
             'RELIANCE',
             'HDFC',
             'ICICIBANK',
             'INFY',
             'TCS',
             'KOTAKBANK',
             'ITC',
             'AXISBANK',
            'SBIN']
```

```python
shutil.copy( location +years[0]+'/'+common+months[0]+years[0]+'/'+companies[0]+'.txt',location)
```

##### Data classifying <B> MONTH WISE</B> into specific <b> YEARS FOLDERS</b>  with name of company

```python

for year in years:     #all years in data
    print('in %s year.................\n'%year)
    for month in months:   #all months data with prefix of nifty50_
        print('in %s month.....................\n'%month)
        for company in companies:  # company in companies list
            #print('for %s company \n'%company)
            source = location + year+'/'+ common +month+year+'/'+company+'.txt' # source for files
            
            parent_dir = location  
            year2 = year+'_data'   #2017_data,2018_data
            path = os.path.join(parent_dir, year2)   #path  name = year name
            path2 = os.path.join(path,company)       #company in year
            
            try:                 #checking for name exist error                
                os.mkdir(path)  # Create the directory with path name(year name)
                os.mkdir(path2)  ## Create the directory with path name(company name)
                print("Directory '% s' created.............../n" % year)
                print("Directory '% s' created........." % company)
            except:
                try:
                    os.mkdir(path2) #check for existed company
                    print("Directory '% s' created............" % company)
                except:
                    print()  #print('----------------------------')
            try:
                shutil.copy(source, path2)
                os.rename(path2+'/'+company+'.txt', path2+'/'+company+'_'+month+'.txt') #HDFC.txt ---> HDFC_month.txt
                print(year+'.....'+company+'_'+month+'.txt'+'/n')
            except:
                print('no file...... %s',month)
            
            
            
#shutil.copyfile(source, destination, *, follow_symlinks = True)
```
